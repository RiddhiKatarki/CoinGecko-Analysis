Crypto Analysis using CoinGecko API

```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
import pandas as pd
```

```python
cg.ping()
```

```python
coin_list = cg.get_coins_list()
print(coin_list) #Get list of all coins listed
```

```python
coin_list_df = pd.DataFrame(coin_list,columns = ['name'])
print(coin_list_df)
```

```python
defi_coins_list = ['1inch','aave','bancor','yearn-finance','uniswap'] # List of DeFi coins
```

```python
cg.get_price(ids = defi_coins_list,vs_currencies = 'usd') #Fetch the latest price of all DeFi coins
```

```python
cg.get_exchanges_list() #Fetch the detials of all exchanges listed
```

```python
print("Binance-US")
cg.get_exchanges_volume_chart_by_id('binance_us','7 days')
#Fetch volume data for Binance-US
```

```python
print("BitFinex")
cg.get_exchanges_volume_chart_by_id('bitfinex','30 days')
#Fetch volume data for Bitfinex

```

```python
print("CoinBase")
cg.get_exchanges_volume_chart_by_id('gdax','30 days')
#Fetch volume data for Coinbase
```
